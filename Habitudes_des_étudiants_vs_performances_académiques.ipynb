# ***A-ESTIMATION DU SCORE DE L'EXAMEN***

# ***test de shapiro-wilk***

# ***H0 : La distribution suit la loi normale***
# ***H1 : La distribution ne suit pas la loi normale***

```python
from scipy import stats

# Perform the Shapiro-Wilk test on the 'exam_score' column
shapiro_test = stats.shapiro(df['exam_score'])

print(f"Shapiro-Wilk test for exam_score:")
print(f"Test Statistic: {shapiro_test.statistic:.4f}")
print(f"P-value: {shapiro_test.pvalue:.4f}")

# Interpret the result
alpha = 0.05
if shapiro_test.pvalue > alpha:
    print("La distribution suit la loi normale (On ne peut rejetter H0).")
else:
    print("La distribution ne suit pas la loi normale (On rejette H0).")
```

# ***estimation***

```python
import numpy as np

# Bootstrap estimation for the mean of 'exam_score'
n_bootstrap_samples = 1000  # Number of bootstrap samples
sample_size = len(df['exam_score'])
bootstrap_means = []

for _ in range(n_bootstrap_samples):
    # Resample with replacement
    bootstrap_sample = df['exam_score'].sample(n=sample_size, replace=True)
    # Calculate the mean of the bootstrap sample
    bootstrap_mean = bootstrap_sample.mean()
    bootstrap_means.append(bootstrap_mean)

# Calculate the 95% confidence interval for the mean
confidence_interval = np.percentile(bootstrap_means, [2.5, 97.5])
print(f"la moyenne du score de l'examen: {df['exam_score'].mean()}")
print("\nBootstrap Confidence Interval (95%) for the mean of 'exam_score':")
print(f"Lower bound: {confidence_interval[0]:.4f}")
print(f"Upper bound: {confidence_interval[1]:.4f}")
```

# ***Le score de l'examen depend il de l'age ?***


```python
import matplotlib.pyplot as plt
from statsmodels.formula.api import ols
# Perform linear regression for 'pourcentage_presence' and 'exam_score'
model = ols('exam_score ~ age', data=df).fit()

# Get the regression equation coefficients
intercept = model.params['Intercept']
slope = model.params['age']

# Create the scatter plot with regression line
plt.figure(figsize=(10, 6))
sns.regplot(x='age', y='exam_score', data=df, scatter=True, line_kws={'color':'red'})

# Add the regression equation to the plot
equation = f'exam_score = {intercept:.2f} + {slope:.2f} * age'
plt.text(df['age'].min(), df['exam_score'].max(), equation, horizontalalignment='left', size='medium', color='red', weight='semibold')

plt.title('Nuage de Points et Ligne de Régression: Score de l\'Examen vs age')
plt.xlabel('age')
plt.ylabel('Score de l\'Examen')
plt.grid(True)
plt.show()

# Perform the Shapiro-Wilk test on the 'exam_score' column
shapiro_test = stats.shapiro(df['exam_score'])

print(f"Shapiro-Wilk test for exam_score:")
print(f"Test Statistic: {shapiro_test.statistic:.4f}")
print(f"P-value: {shapiro_test.pvalue:.4f}")

# Interpret the result
alpha = 0.05
if shapiro_test.pvalue > alpha:
    print("La distribution suit la loi normale (On ne peut rejeter H0).")
else:
    print("La distribution ne suit pas la loi normale (On rejette H0).")

# Perform the Shapiro-Wilk test on the 'age' column
shapiro_test = stats.shapiro(df['age'])

print(f"Shapiro-Wilk test for age:")
print(f"Test Statistic: {shapiro_test.statistic:.4f}")
print(f"P-value: {shapiro_test.pvalue:.4f}")

# Interpret the result
alpha = 0.05
if shapiro_test.pvalue > alpha:
    print("La distribution suit la loi normale (On ne peut rejeter H0).")
else:
    print("La distribution ne suit pas la loi normale (On rejette H0).")

# H0: Il n'y a pas de corrélation entre le pourcentage de presence et le score de l'examen (rho = 0).
# H1: Il existe une corrélation entre le pourcentage de presence et le score de l'examen (rho != 0).
# Perform Spearman's rank correlation test
spearman_corr, spearman_pvalue = stats.spearmanr(df['age'], df['exam_score'])

print("\nTest de Corrélation de Spearman entre Exam_score et age:")
print(f"Coefficient de corrélation de Spearman (rho): {spearman_corr:.4f}")
print(f"P-value: {spearman_pvalue:.4f}")

# Interpret the result
alpha = 0.05
if spearman_pvalue < alpha:
    print("On rejette H0. Il existe une liaison significative entre age et le score de l'examen.")
else:
    print("On ne rejette pas H0. Il n'y a pas de preuve suffisante d'une liaison significative entre age et le score de l'examen.")
```

# ***Le score de l'exam depend il du sexe ?***

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
sns.boxplot(x='Sexe', y='exam_score', data=df, palette='viridis')
plt.title('Boxplot du Score de l\'Examen par Sexe')
plt.xlabel('Sexe')
plt.ylabel('Score de l\'Examen')
plt.show()

# --- Tests de liaison ---

# 1ère étape : Test de Shapiro-Wilk pour chaque groupe de travail à temps partiel
part_time_groups = df['Sexe'].unique()

print("Test de Shapiro-Wilk pour 'exam_score' par groupe de 'Sexe':")
for group in part_time_groups:
    subset = df[df['Sexe'] == group]['exam_score']
    shapiro_test_group = stats.shapiro(subset)

    print(f"\nGroupe de Sexe: {group}")
    print(f"  Test Statistic: {shapiro_test_group.statistic:.4f}")
    print(f"  P-value: {shapiro_test_group.pvalue:.4f}")

    alpha = 0.05
    if shapiro_test_group.pvalue > alpha:
        print("  La distribution semble suivre la loi normale (On ne peut rejeter H0).")
    else:
        print("  La distribution ne semble pas suivre la loi normale (On rejette H0).")

# 2ème étape : tester l’égalité des variances (Test de Levene)
groups_part_time = [df['exam_score'][df['Sexe'] == group] for group in df['Sexe'].unique()]
stat_levene, p_levene = stats.levene(*groups_part_time)
print(f"\nLevene test: Statistic={stat_levene:.5f}, P-value={p_levene:.5f}")
if p_levene > 0.05:
  print(f"On ne peut rejeter H0. Les variances sont similaires.")
else:
  print(f"On peut rejeter H0. Les variances sont significativement différentes.")

# 3ème étape : tester l’égalité des moyennes (ANOVA si normalité et égalité des variances, sinon Kruskal-Wallis)

# For Kruskal-Wallis, group the data
grouped_data_part_time = [df['exam_score'][df['Sexe'] == group] for group in df['Sexe'].unique()]

# Test de Kruskal-Wallis
kruskal_result_part_time = stats.kruskal(*grouped_data_part_time)
print("\nKruskal-Wallis results for exam_score vs Sexe:")
print(f"Statistic: {kruskal_result_part_time.statistic:.4f}, P-value: {kruskal_result_part_time.pvalue:.4f}")

# Interprétation :
alpha = 0.05
if kruskal_result_part_time.pvalue < alpha:
    print("→ Il y a une liaison significative entre le score de l'examen et le Sexe (p < 0.05).")
else:
    print("→ Il n'y a pas de preuve suffisante d'une liaison significative entre le score de l'examen et le Sexe (p ≥ 0.05).")
```

```python
import matplotlib.pyplot as plt
# Perform linear regression for 'study_hours_per_day' and 'exam_score'
model = ols('exam_score ~ heure_etud_jour', data=df).fit()

# Get the regression equation coefficients
intercept = model.params['Intercept']
slope = model.params['heure_etud_jour']

# Create the scatter plot with regression line
plt.figure(figsize=(10, 6))
sns.regplot(x='heure_etud_jour', y='exam_score', data=df, scatter=True, line_kws={'color':'red'})

# Add the regression equation to the plot
equation = f'exam_score = {intercept:.2f} + {slope:.2f} * heure_etud_jour'
plt.text(df['heure_etud_jour'].min(), df['exam_score'].max(), equation, horizontalalignment='left', size='medium', color='red', weight='semibold')

plt.title('Nuage de Points et Ligne de Régression: Score de l\'Examen vs Heures d\'Étude par Jour')
plt.xlabel('Heures d\'Étude par Jour')
plt.ylabel('Score de l\'Examen')
plt.grid(True)
plt.show()
# Perform the Shapiro-Wilk test on the 'exam_score' column
shapiro_test = stats.shapiro(df['exam_score'])

print(f"Shapiro-Wilk test for exam_score:")
print(f"Test Statistic: {shapiro_test.statistic:.4f}")
print(f"P-value: {shapiro_test.pvalue:.4f}")

# Interpret the result
alpha = 0.05
if shapiro_test.pvalue > alpha:
    print("La distribution suit la loi normale (On ne peut rejeter H0).")
else:
    print("La distribution ne suit pas la loi normale (On rejette H0).")

# Perform the Shapiro-Wilk test on the 'heure_etud_jour' column
shapiro_test = stats.shapiro(df['heure_etud_jour'])

print(f"Shapiro-Wilk test for heure_etud_jour:")
print(f"Test Statistic: {shapiro_test.statistic:.4f}")
print(f"P-value: {shapiro_test.pvalue:.4f}")

# Interpret the result
alpha = 0.05
if shapiro_test.pvalue > alpha:
    print("La distribution suit la loi normale (On ne peut rejetter H0).")
else:
    print("La distribution ne suit pas la loi normale (On rejette H0).")

# H0: Il n'y a pas de corrélation entre les heures d'étude par jour et le score de l'examen (rho = 0).
# H1: Il existe une corrélation entre les heures d'étude par jour et le score de l'examen (rho != 0).

# Perform Spearman's rank correlation test
spearman_corr, spearman_pvalue = stats.spearmanr(df['heure_etud_jour'], df['exam_score'])

print("\nTest de Corrélation de Spearman entre Exam_score et Heures d'Étude par Jour:")
print(f"Coefficient de corrélation de Spearman (rho): {spearman_corr:.4f}")
print(f"P-value: {spearman_pvalue:.4f}")

# Interpret the result
alpha = 0.05
if spearman_pvalue < alpha:
    print("On rejette H0. Il existe une liaison significative entre les heures d'étude par jour et le score de l'examen.")
else:
    print("On ne rejette pas H0. Il n'y a pas de preuve suffisante d'une liaison significative entre les heures d'étude par jour et le score de l'examen.")

import numpy as np
from scipy import stats

def custom_correlation(x, y, theo=0.8116):
 # Calculer la corrélation de Pearson et la p-value
 corr_coeff, p_value = stats.spearmanr(x, y)

 # Comparer la corrélation obtenue avec la valeur théorique
 diff = corr_coeff - theo

 return {
 'correlation_coefficient': corr_coeff,
 'p_value': p_value,
 'difference_from_theoretical': diff
 }
# Test de la fonction
result = custom_correlation(df['heure_etud_jour'], df['exam_score'], theo=0.5)
print(f"Coefficient de corrélation de Spearman: {result['correlation_coefficient']:.4f}")
print(f"P-value: {result['p_value']:.4f}")
if result['p_value'] > 0.05:
    print("Le coefficient de corrélation est significativement egale à 0.8116.")
else:
    print("Le coefficient de corrélation est significativement different de 0.8116.")

import scipy.stats as stats

# Supposons que df est un DataFrame contenant 'heure_etud_jour' et 'exam_score'
corr, p_value = stats.spearmanr(df['heure_etud_jour'], df['exam_score'], alternative='greater')

print(f"Spearman correlation coefficient: {corr:.4f}")
print(f"P-value: {p_value:.4f}")

# Interprétation :
if p_value < 0.05:
    print("→ On rejette H0 : Il existe une corrélation positive significative.")
else:
    print("→ On ne peut pas rejeter H0 : Pas de corrélation positive significative.")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from prince import FAMD

# Charger vos données ici si nécessaire
# df = pd.read_csv('votre_fichier.csv')

# Exemple fictif de DataFrame (à remplacer par votre dataset)
# df = pd.DataFrame({...})

# Étape 1 : Sélectionner les colonnes pertinentes
if 'student_id' in df.columns:
    variables_famd = df.drop('student_id', axis=1).columns
else:
    variables_famd = df.columns

df_famd = df[variables_famd].copy()

# Étape 2 : Gérer les valeurs manquantes
if df_famd.isnull().any().any():
    print("Attention : Des valeurs manquantes ont été détectées. Imputation en cours...")
    for col in df_famd.columns:
        if pd.api.types.is_categorical_dtype(df_famd[col]) or df_famd[col].dtype == 'object':
            df_famd[col] = df_famd[col].fillna(df_famd[col].mode()[0])
        elif pd.api.types.is_numeric_dtype(df_famd[col]):
            df_famd[col] = df_famd[col].fillna(df_famd[col].mean())

# Étape 3 : Convertir les colonnes object en catégorie
for col in df_famd.columns:
    if df_famd[col].dtype == 'object':
        df_famd[col] = df_famd[col].astype('category')

# Étape 4 : Vérifier qu'il y a assez de variables
n_cols = df_famd.shape[1]
if n_cols <= 1:
    raise ValueError("Erreur : FAMD nécessite au moins 2 variables.")

# Étape 5 : Définir le nombre de composantes
n_components = min(n_cols - 1, 10)

# Étape 6 : Appliquer FAMD
famd = FAMD(
    n_components=n_components,
    n_iter=3,
    random_state=42
)
famd.fit(df_famd)

# Étape 7 : Récupérer les valeurs propres et l'inertie expliquée
eigenvalues = famd.eigenvalues_
explained_inertia = eigenvalues / sum(eigenvalues)
cumulative_inertia = explained_inertia.cumsum()

# Afficher les valeurs propres et l'inertie
print("\nValeurs propres et inertie expliquée :")
eigen_df = pd.DataFrame({
    'Valeur propre': eigenvalues,
    'Inertie expliquée (%)': explained_inertia * 100,
    'Inertie cumulée (%)': cumulative_inertia * 100
})
print(eigen_df)

# Étape 8 : Graphique du coude (Elbow plot)
plt.figure(figsize=(8, 5))
plt.plot(range(1, len(eigenvalues) + 1), eigenvalues, marker='o', linestyle='-')
plt.title("Graphique des valeurs propres (critère du coude)")
plt.xlabel("Dimension")
plt.ylabel("Valeur propre")
plt.grid(True)
plt.show()

# Étape 9 : Coordonnées des individus
individual_coords = famd.row_coordinates(df_famd)
individual_coords.head()

# Étape 10 : Coordonnées des variables (qualitatives et quantitatives)
all_variable_coords = famd.column_coordinates_

# Séparer les types de variables
quali_vars = df_famd.select_dtypes(include=['category']).columns
quanti_vars = df_famd.select_dtypes(include=[np.number]).columns

variable_coords_quali = all_variable_coords.loc[all_variable_coords.index.get_level_values(0).isin(quali_vars)]
variable_loadings_quanti = all_variable_coords.loc[quanti_vars]

print("\nCoordonnées des catégories qualitatives :")
print(variable_coords_quali.head())

print("\nCorrélations des variables quantitatives avec les dimensions :")
print(variable_loadings_quanti.head())

# Étape 11 : Visualisation des individus et des variables
if individual_coords.shape[1] >= 2:
    fig, axs = plt.subplots(1, 2, figsize=(16, 8))

    # Nuage des individus
    axs[0].scatter(individual_coords.iloc[:, 0], individual_coords.iloc[:, 1], alpha=0.6)
    axs[0].set_title("Individus sur les deux premières dimensions")
    axs[0].set_xlabel(f"Dim 1 ({explained_inertia[0]*100:.1f}%)")
    axs[0].set_ylabel(f"Dim 2 ({explained_inertia[1]*100:.1f}%)")
    axs[0].grid(True)

    # Nuage des variables
    for name, coords in variable_loadings_quanti.iterrows():
        axs[1].arrow(0, 0, coords[0], coords[1], head_width=0.03, length_includes_head=True, color='b')
        axs[1].text(coords[0], coords[1], name, color='b', ha='center', va='center')

    for name, coords in variable_coords_quali.iterrows():
        axs[1].scatter(coords[0], coords[1], color='r')
        axs[1].text(coords[0], coords[1], name, color='r', ha='center', va='center')

    axs[1].set_title("Variables sur les deux premières dimensions")
    axs[1].set_xlabel(f"Dim 1 ({explained_inertia[0]*100:.1f}%)")
    axs[1].set_ylabel(f"Dim 2 ({explained_inertia[1]*100:.1f}%)")
    axs[1].axhline(0, color='black', linestyle='--', linewidth=0.8)
    axs[1].axvline(0, color='black', linestyle='--', linewidth=0.8)
    axs[1].grid(True)
    axs[1].set_xlim(-1.2, 1.2)
    axs[1].set_ylim(-1.2, 1.2)
    axs[1].set_aspect('equal')

    plt.tight_layout()
    plt.show()

elif individual_coords.shape[1] == 1:
    print("INFO : Seulement 1 dimension extraite. Affichage 1D.")
    fig, axs = plt.subplots(1, 2, figsize=(16, 4))

    axs[0].scatter(individual_coords.iloc[:, 0], np.zeros_like(individual_coords), alpha=0.6)
    axs[0].set_title("Individus sur la première dimension")
    axs[0].set_xlabel(f"Dim 1 ({explained_inertia[0]*100:.1f}%)")
    axs[0].set_yticks([])
    axs[0].grid(True)

    for name, coords in variable_loadings_quanti.iterrows():
        axs[1].scatter(coords[0], 0, color='b', marker='x')
        axs[1].text(coords[0], 0.05, name, color='b', ha='center')

    for name, coords in variable_coords_quali.iterrows():
        axs[1].scatter(coords[0], 0, color='r')
        axs[1].text(coords[0], -0.05, name, color='r', ha='center')

    axs[1].set_title("Variables sur la première dimension")
    axs[1].set_xlabel(f"Dim 1 ({explained_inertia[0]*100:.1f}%)")
    axs[1].set_yticks([])
    axs[1].grid(True)
    axs[1].axhline(0, color='black', linestyle='--')

    plt.tight_layout()
    plt.show()
else:
    print("Aucune dimension extraite. Impossible d'afficher les graphiques.")
```

```python
# CLUSTERING HIERARCHIQUE ASCENDANT (CAH)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage

# Étape 1: Préparer les données pour le clustering
# On utilise les coordonnées des individus après FAMD
data_for_cah = individual_coords.values  # Utilisation des coordonnées factorielles issues de FAMD

# Étape 2: Calculer la matrice de liaison (utilisation de la méthode 'ward')
Z = linkage(data_for_cah, method='ward')

# Étape 3: Afficher le dendrogramme
plt.figure(figsize=(15, 8))
plt.title('Dendrogramme de la CAH')
plt.xlabel('Index des individus')
plt.ylabel('Distance euclidienne (méthode Ward)')
dendrogram(
    Z,
    leaf_rotation=90.,  # Pivoter les étiquettes des feuilles
    leaf_font_size=8.,  # Ajuster la taille de la police des feuilles
    show_contracted=True, # Affiche les contractions (branches fusionnées)
)
plt.show()

# Étape 4: Choisir le nombre de clusters et appliquer le clustering
# Exemple avec 3 clusters
n_clusters = 3
model_cah = AgglomerativeClustering(n_clusters=n_clusters, linkage='ward')  # 'affinity' retiré
df['cluster_cah'] = model_cah.fit_predict(data_for_cah)

# Étape 5: Analyser les clusters
print(f"\nRépartition des individus par cluster ({n_clusters} clusters) :")
print(df['cluster_cah'].value_counts())

print("\nDescription des clusters par la moyenne des variables quantitatives :")
print(df.groupby('cluster_cah')[quanti_vars].mean())

print("\nDescription des clusters par la distribution des variables qualitatives :")
for col in quali_vars:
    print(f"\nDistribution de '{col}' par cluster :")
    print(df.groupby('cluster_cah')[col].value_counts(normalize=True).unstack().fillna(0).round(2))

# Visualiser les clusters sur les deux premières dimensions du FAMD
if individual_coords.shape[1] >= 2:
    plt.figure(figsize=(10, 8))
    sns.scatterplot(
        x=individual_coords.iloc[:, 0],
        y=individual_coords.iloc[:, 1],
        hue=df['cluster_cah'],
        palette='viridis',
        legend='full'
    )
    plt.title(f'Individus regroupés en {n_clusters} clusters sur les dimensions FAMD 1 et 2')
    plt.xlabel(f"Dim 1 ({explained_inertia[0]*100:.1f}%)")
    plt.ylabel(f"Dim 2 ({explained_inertia[1]*100:.1f}%)")
    plt.grid(True)
    plt.show()
```

```python
# Analyze clusters to assign themes
print("\nAnalyse et Thèmes des Clusters :")
print("-" * 40)

# Get cluster centers for quantitative variables
quanti_cluster_means = df.groupby('cluster_cah')[quanti_vars].mean()
print("\nMoyenne des variables quantitatives par cluster:")
print(quanti_cluster_means)

# Get cluster distributions for qualitative variables
quali_cluster_distributions = {}
for col in quali_vars:
    quali_cluster_distributions[col] = df.groupby('cluster_cah')[col].value_counts(normalize=True).unstack().fillna(0).round(2)
    print(f"\nDistribution de '{col}' par cluster:")
    print(quali_cluster_distributions[col])


# --- Assign Themes Based on Analysis ---
# This part requires manual interpretation based on the printed outputs above.
# You look at which variables (quantitative means and qualitative distributions)
# are significantly higher or lower in each cluster compared to the overall mean or distribution.

themes = {}
n_clusters = df['cluster_cah'].nunique()

print("\nProposition de Thèmes pour Chaque Cluster :")
print("-" * 40)

for cluster_id in range(n_clusters):
    print(f"\n--- Cluster {cluster_id} ---")
    cluster_theme_description = f"Cluster {cluster_id}: "

    # Analyze quantitative variables for this cluster
    print("Caractéristiques quantitatives:")
    for var in quanti_vars:
        cluster_mean = quanti_cluster_means.loc[cluster_id, var]
        overall_mean = df[var].mean()
        diff_percentage = ((cluster_mean - overall_mean) / overall_mean) * 100 if overall_mean != 0 else (cluster_mean - overall_mean) * 100
        print(f"  - {var}: Moyenne = {cluster_mean:.2f} (vs globale = {overall_mean:.2f}, différence = {diff_percentage:.1f}%)")

    # Analyze qualitative variables for this cluster
    print("Caractéristiques qualitatives:")
    for var in quali_vars:
        if var in quali_cluster_distributions:
            print(f"  - {var}:")
            cluster_dist = quali_cluster_distributions[var].loc[cluster_id]
            overall_dist = df[var].value_counts(normalize=True).round(2)
            for category, percentage in cluster_dist.items():
                 overall_percentage = overall_dist.get(category, 0)
                 diff_percentage_points = percentage - overall_percentage
                 print(f"    - {category}: {percentage:.2f} (vs globale = {overall_percentage:.2f}, différence = {diff_percentage_points:.2f} points)")

    # Based on the analysis above, formulate a theme. This is subjective and depends
    # on which variables show the most significant differences for the cluster.

    # Example themes (replace with your actual interpretations)
    if cluster_id == 0:
        # Example interpretation: High exam scores, high attendance, high study hours, maybe certain internet quality or parent education levels prominent
        themes[cluster_id] = "Étudiants Performants et Assidus"
    elif cluster_id == 1:
         # Example interpretation: Lower exam scores, more time on social media/netflix, lower study hours, potentially lower parent education or internet quality
        themes[cluster_id] = "Étudiants avec Moins de Performance et Davantage d'Activités Extra-Académiques"
    elif cluster_id == 2:
         # Example interpretation: Average characteristics, maybe slightly different patterns in sleep or exercise.
        themes[cluster_id] = "Étudiants Moyens ou avec Habitudes Spécifiques"
    # Add more elif blocks if you have more clusters

    if cluster_id in themes:
        print(f"\nThème proposé pour le Cluster {cluster_id}: {themes[cluster_id]}")
    else:
        print(f"\nThème à déterminer pour le Cluster {cluster_id} basé sur l'analyse ci-dessus.")

# You can store the themes in the DataFrame or use them for further analysis
df['cluster_theme'] = df['cluster_cah'].map(themes)
print("\nDataFrame avec les Thèmes des Clusters ajoutés:")
print(df[['cluster_cah', 'cluster_theme']].head())

```

# ****h-test de chow****

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from scipy import stats

# 🔧 Fonction pour le test de Chow
def chow_test(X, y, last_index):
    """
    Effectue le test de Chow pour détecter une rupture structurelle à un point donné.

    Parameters:
    - X: DataFrame ou array des variables explicatives
    - y: Series ou array, variable cible
    - last_index: int, position de fin du premier sous-groupe (rupture)

    Returns:
    - F_stat: Valeur de la statistique F
    - p_value: P-value associée
    """

    # Convertir en numpy arrays si nécessaire
    X = sm.add_constant(X)  # Ajouter la constante

    # Taille totale
    n = len(X)
    k = X.shape[1]  # Nombre de coefficients (avec intercept)

    # Diviser en deux groupes
    X1, X2 = X[:last_index], X[last_index:]
    y1, y2 = y[:last_index], y[last_index:]

    # Vérifier qu'il y a assez d'observations dans chaque groupe
    if len(y1) < k or len(y2) < k:
        raise ValueError("Pas assez d'observations dans un des deux sous-groupes.")

    # Modèle complet
    model_full = sm.OLS(y, X).fit()
    RSS_full = model_full.ssr

    # Modèle groupe 1
    model1 = sm.OLS(y1, X1).fit()
    RSS1 = model1.ssr

    # Modèle groupe 2
    model2 = sm.OLS(y2, X2).fit()
    RSS2 = model2.ssr

    # Statistique du test de Chow
    numerator = (RSS_full - (RSS1 + RSS2)) / k
    denominator = (RSS1 + RSS2) / (n - 2*k)
    F_stat = numerator / denominator
    p_value = 1 - stats.f.cdf(F_stat, k, n - 2*k)

    return F_stat, p_value


# 📊 Données exemple
X = data[['heure_etud_jour', 'heure_reseaux_sociaux', 'heure_netflix',
          'pourcentage_presence', 'heure_sommeil', 'frequence_exercice',
          'qualite_internet', 'evaluation_sante_mental']]  # Variables explicatives
y = data['exam_score']  # Variable cible
break_point = len(data) // 2  # Rupture au milieu

# 🚀 Effectuer le test de Chow
F_stat, p_value = chow_test(X, y, break_point)

# 📈 Affichage des résultats
print(f"Statistique F : {F_stat:.4f}")
print(f"P-value : {p_value:.4f}")

if p_value < 0.05:
    print("=> Rejet de H₀ : il y a une rupture de structure.")
else:
    print("=> Non rejet de H₀ : pas de rupture détectée.")
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.neural_network import MLPRegressor # Import MLPRegressor

# Assurez-vous que 'a' est votre DataFrame contenant les données
# 'exam_score' est la variable cible

# Définir les variables indépendantes (X) et la variable dépendante (y)
X = a.drop('exam_score', axis=1)
y = a['exam_score']

# Initialiser les modèles
models = {
    "Regression Linéaire": LinearRegression(),
    "Ridge Regression": Ridge(),
    "Lasso Regression": Lasso(),
    "Elastic Net": ElasticNet(),
    "K-Neighbors Regressor": KNeighborsRegressor(),
    "Decision Tree Regressor": DecisionTreeRegressor(random_state=42),
    "Random Forest Regressor": RandomForestRegressor(random_state=42),
    "Gradient Boosting Regressor": GradientBoostingRegressor(random_state=42),
    "AdaBoost Regressor": AdaBoostRegressor(random_state=42),
    "Support Vector Regressor (Linear)": SVR(kernel='linear'),
    "Support Vector Regressor (RBF)": SVR(kernel='rbf'),
    "MLP Regressor": MLPRegressor(random_state=42, max_iter=1000) # Increased max_iter
}

# Nombre de folds pour la validation croisée K-Fold
n_splits = 5 # Vous pouvez ajuster le nombre de folds

# Initialiser K-Fold
kf = KFold(n_splits=n_splits, shuffle=True, random_state=42)

# Dictionnaire pour stocker les scores de chaque modèle sur chaque fold
model_scores = {name: {'rmse': [], 'mae': [], 'r2': []} for name in models.keys()}

# Boucle sur chaque modèle
for name, model in models.items():
    print(f"\nValidation croisée K-Fold pour : {name}")
    fold_count = 1
    # Boucle sur chaque fold
    for train_index, test_index in kf.split(X):
        X_train, X_test = X.iloc[train_index], X.iloc[test_index]
        y_train, y_test = y.iloc[train_index], y.iloc[test_index]

        try:
            # Entraîner le modèle
            model.fit(X_train, y_train)

            # Faire des prédictions sur l'ensemble de test du fold
            y_pred = model.predict(X_test)

            # Calculer les métriques pour ce fold
            rmse = np.sqrt(mean_squared_error(y_test, y_pred))
            mae = mean_absolute_error(y_test, y_pred)
            r2 = r2_score(y_test, y_pred)

            # Stocker les scores pour ce fold
            model_scores[name]['rmse'].append(rmse)
            model_scores[name]['mae'].append(mae)
            model_scores[name]['r2'].append(r2)

            # print(f"  Fold {fold_count}: RMSE={rmse:.4f}, MAE={mae:.4f}, R²={r2:.4f}")
            fold_count += 1

        except Exception as e:
            print(f"  Erreur dans Fold {fold_count} pour {name}: {e}")
            # Ajouter NaN pour ce fold si une erreur se produit
            model_scores[name]['rmse'].append(np.nan)
            model_scores[name]['mae'].append(np.nan)
            model_scores[name]['r2'].append(np.nan)
            fold_count += 1


# Calculer les métriques moyennes sur tous les folds
average_results = []
for name, scores in model_scores.items():
    avg_rmse = np.nanmean(scores['rmse']) # Use nanmean to handle potential NaNs from errors
    avg_mae = np.nanmean(scores['mae'])
    avg_r2 = np.nanmean(scores['r2'])

    # Calculer le R-carré ajusté (applicable principalement à la régression linéaire)
    # Pour d'autres modèles, le R2 standard est plus pertinent.
    # Ici, pour une comparaison uniforme, on utilise le R2 moyen.
    # Si une analyse approfondie du R2 ajusté est nécessaire, elle doit être faite séparément pour LinearRegression.
    # Formule R2 ajusté: 1 - (1 - R2) * (n-1) / (n-p-1)
    # n = nombre d'observations (total)
    # p = nombre de prédicteurs
    # Le calcul du R2 ajusté moyen sur les folds est complexe et moins standard
    # On se basera sur le R2 moyen ici pour la comparaison.

    average_results.append({'Modèle': name, 'RMSE Moy.': avg_rmse, 'MAE Moy.': avg_mae, 'R² Moy.': avg_r2})

# Convertir les résultats moyens en DataFrame
average_results_df = pd.DataFrame(average_results)

# Triez par une métrique de performance moyenne (par exemple, RMSE, du plus bas au plus haut)
average_results_df_sorted = average_results_df.sort_values(by='RMSE Moy.', ascending=True)

print("\n--- Tableau Récapitulatif des Métriques Moyennes (Validation Croisée K-Fold) ---")
print(average_results_df_sorted.round(4))

# Identifier les 3 meilleurs modèles basés sur le RMSE moyen
top_3_models_kfold = average_results_df_sorted.head(3)
```
